```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
features = pd.read_csv('../../data/222k_dinov2_features.csv')
```

```python
map = {"Unnamed: 0" : "species", "Unnamed: 1" : "filename"}
features = features.rename(columns=map)
features['filename'] = features['filename'].str.replace('.txt', '.png', regex=False)
features
```

```python
tsv_file = "../../data/ecotaxa_export_10210_20241126_0935.tsv"
tsv_df = pd.read_csv(tsv_file, sep='\t')
tsv_df
```

```python
desired_col = "object_annotation_category"


def add_annotation_category(tsv_df, dataframe, filename_column='filename', img_file_name_column='img_file_name', annotation_column='object_annotation_category'):
    """
    Matches the `img_file_name` in the TSV file with the `filename` column in the DataFrame
    and adds the `object_annotation_category` to the DataFrame.

    Parameters:
    tsv_file (str): Path to the TSV file.
    dataframe (pd.DataFrame): The DataFrame with a `filename` column.
    filename_column (str): The column name in the DataFrame to match with `img_file_name` in TSV.
    img_file_name_column (str): The column name in the TSV for matching.
    annotation_column (str): The column name in the TSV for the annotation category.

    Returns:
    pd.DataFrame: The updated DataFrame with an added `object_annotation_category` column.
    """
    # Merge the DataFrame with the TSV data on the corresponding filename columns
    tsv_df[img_file_name_column] = tsv_df[img_file_name_column].apply(os.path.basename)
    merged_df = dataframe.merge(
        tsv_df[[img_file_name_column, annotation_column]],
        how = 'left',
        left_on = filename_column,
        right_on = img_file_name_column
    )
    
    # Drop the redundant column added by the merge (if necessary)
    merged_df.drop(columns=[img_file_name_column], inplace=True)
    
    return merged_df
```

```python
new_features = add_annotation_category(tsv_df, features)
new_features.pop('species')

last_column = new_features.columns[-1]

# Rearrange the columns: bring the last column to the front
df = new_features[[last_column] + [col for col in new_features.columns if col != last_column]]
```

```python
df.to_csv('../../data/222k_all_dino2_features.csv')
```
